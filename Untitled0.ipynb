2) Create tasks like generation, summarization, Q&A, translation.

```python
from transformers import pipeline
text_generation = pipeline("text-generation", model = "gpt2")
text_summarization = pipeline("summarization",model = "facebook/bart-large-cnn")
text_question = pipeline("question-answering",model = "distilbert-base-uncased-distilled-squad")
text_translation = pipeline("translation_en_to_fr", model = "Helsinki-NLP/opus-mt-en-fr")

print(text_generation("what is life?"))
print(text_summarization("The internet has revolutionized the way we communicate and access information.People can now connect instantly across the world through social media and messaging apps."))
print(text_question(question="who is primeminister of india?", context="Narendra Modi is the current prime minister of India."))
print(text_translation("hello!, Good morning"))
```

3) Try few-shot, zero-shot, and chain-of-thought prompting.
   1. zero-shot
   

```python
from transformers import pipeline
translator = pipeline("translation_en_to_te", model="Meher2006/english-to-telugu-model")
text = ("The internet has revolutionized the way we communicate and access information. "
        "People can now connect instantly across the world through social media and messaging apps. "
        "It has also opened up countless opportunities for online learning and remote work.")
english_text = "This is a test sentence to translate from English to Telugu."
telugu_translation = translator(english_text)
print(telugu_translation[0]['translation_text'])
# ➤ Analysis:
# - Works decently with just the instruction.
# - Captures the main idea, but output may vary in wording.
# - No examples to guide tone or length.
```

2. few short

```python
few_shot_prompt = (
    "Text: The sun is the closest star to Earth. It gives us light and heat.\n"
    "Summary: The sun provides Earth with light and warmth.\n"
    f"Text: {text}\n"
    "Summary:"
)
print("Few-shot summary:\n", translator(few_shot_prompt))
# ➤ Analysis:
# - More focused and concise than zero-shot.
# - Learns format and tone from the example.
# - More likely to match expected summary style.
```

3. chain-of-thought

```python
cot_prompt = (
    "Summarize the paragraph below.\n"
    "Step 1: Identify the main topic.\n"
    "Step 2: Pick out key ideas about communication and remote opportunities.\n"
    "Step 3: Write a short summary based on these points.\n"
    f"Text: {text}\n"
    "Summary:"
)
print("Chain-of-thought summary:\n", translator(cot_prompt))
# ➤ Analysis:
# - Encourages step-by-step thinking.
# - Slightly more accurate and detailed.
# - Useful for complex tasks but longer prompts.
```

```python
print("\n--- Summary Comparison ---")
print("Zero-shot:", telugu_translation[0])
print("Few-shot summary:\n", translator(few_shot_prompt))
print("Chain-of-thought summary:\n", translator(cot_prompt))

# ➤ Final Comments:
# - Zero-shot is good for quick tasks.
# - Few-shot is best when you can give format examples.
# - Chain-of-Thought works best for logical or multi-step tasks.
```
